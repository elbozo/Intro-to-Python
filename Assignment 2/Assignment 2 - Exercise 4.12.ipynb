## Simulation: The Tortoise and the Hare

```python
import random

T = 0
H = 0

def T_move()->int:
    R = random.randint(1,10)
    if(R in range(1,6)):
        return 3
    elif(R in range(6,8)):
        return -6
    else:
        return 1
def H_move()->int:
    R = random.randint(1,10)
    if(R in range(1,3)):
        return 0
    elif(R in range(3,5)):
        return 9
    elif(R == 5):
        return -12
    elif(R in range(6,9)):
        return 1
    else:
        return -2

print("BANG !!!!!\nAND THEY'RE OFF !!!!!")
while(True):
    for i in range(70):
        if(T==H and i == T):
            print('OUCH!!',end='')
        else:
            if(i==T):
                print('T',end='')
            elif(i==H):
                print('H',end='')
            else:
                print('-',end='')
    print()
    if(T>=70 or H >=70):
        break
    T_next = T_move()
    H_next = H_move()
    T = T + T_next
    H = H + H_next
    if(T<0):
        T = 0
    if(H<0):
        H = 0

if(T>H):
    print('TORTOISE WINS!!! YAY!!!')
elif(T<H):
    print('Hare wins.')
else:
    print('It is a tie')
```
